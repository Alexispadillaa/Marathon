Filtros aplicados

```python
df2 = df[(df["Event distance/length"].isin(['50km', '50mi'])) & (df["Year of event"] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

Eliminar "(USA)" de la columna Event name

```python
df2["Event name"] = df2["Event name"].str.split('(').str.get(0)
```

```python
df2.head()
```

Limpiar edad de atleta

```python
df2["Athlete age"] = 2024 - df2["Athlete year of birth"]
```

Eliminar 'h' al final de athlete performance

```python
df2["Athlete performance"] = df2["Athlete performance"].str.split(' ').str.get(0)
```

```python
df2.head()
```

Eliminar columnas: Athlete club, Athlete countrty, Athlete year of birth, Athlete age category

```python
df2 = df2.drop(["Athlete club", "Athlete country", "Athlete year of birth", "Athlete age category"], axis = 1)
```

```python
df2.head()
```

Limpiar valores nulos

```python
#Revisar cuantos valores nulos hay por columna
df2.isna().sum()
```

```python
#Revisar datos nulos de esas columna(s)
df2[df2["Athlete age"].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.isna().sum()
df2.shape
```

Revisar duplicados

```python
df2[df2.duplicated() == True]
```

```python
df2.reset_index(drop = True)
```

```python
df2["Athlete age"] = df2["Athlete age"].astype(int)
```

```python
df2["Athlete average speed"] = df2["Athlete average speed"].astype(float)
```

```python
df2 = df2.rename(columns = {
    'Year of event': 'Year',
    'Event dates': 'Race day', 
    'Event name': 'Race name',
    'Event distance/length': 'Race length',
    'Event number of finishers': 'Race number of finishers',
    'Athlete performance': 'Athlete performance',
    'Athlete gender': 'Athlete gender',
    'Athlete average speed': 'Athlete average speed',
    'Athlete ID': 'Athlete id'
})
```

# Preguntas que responder con el Dataframe

```python
df3.head()
```

Diferencias en velocidad entre 50km, 50mi en hombre y mujeres

```python
df3.groupby(['Race length', 'Athlete gender'])['Athlete average speed'].mean()
```

¿Cuales son los 15 mejores grupos de edades en 50mi?

```python
df3.query('`Race length` == "50mi"').groupby('Athlete age')['Athlete average speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19').head(15)
```

¿Cuales son los 20 peores grupos de edades en 50mi?

```python
df3.query('`Race length` == "50mi"').groupby('Athlete age')['Athlete average speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count>19').head(20)
```

**Acomodar meses por estaciones del año**

Spring  3-5

Summer 6-8

Fall 9-11

Winter 12-2



```python
df3['Race Month'] = df3["Race day"].str.split(".").str.get(1).astype(int)
```

```python
df3["Race season"] = df3["Race Month"].apply(lambda x: "Spring" if x > 2 and x < 6 else "Summer" if x > 5 and x < 9 else "Fall" if x > 8 and x < 12 else "Winter")
```

```python
m = df3.loc[df3["Race Month"] == 9]
m.head()
```

Encontrar la media de velocidad por estación del año

```python
df3.groupby("Race season")["Athlete average speed"].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

Encontrar la media de velocidad por estación del año usando 50mi únicamente

```python
df3.query("`Race length` == '50mi'").groupby("Race season")["Athlete average speed"].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
